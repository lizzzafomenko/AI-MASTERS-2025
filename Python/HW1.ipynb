По кругу записано несколько букв (возможно, повторяющихся). Студент хочет узнать, сможет ли он прочитать некоторое слово, если будет двигаться по кругу (в любом направлении), не пропуская буквы. Студент сам выбирает место, с которого он начинает читать, и направление. Необходимо написать функцию find_word_in_circle(circle, word), которая должна возвращать позицию, с которой нужно начинать чтение (индекс в строке) и направление чтения (1 если слева направо или -1 если справа налево), если студент может найти строку word в круговой строке circle, и -1 иначе. Строка word содержит как минимум один символ.

Формат ввода

find_word_in_circle(’napo’, ’ap’)

Формат вывода

1 1

Примечания

На проверку необходимо отправить файл с функцией find_word_in_circle(circle, word)



````python
def find_word_in_circle(circle, word):
    if not circle or not word:
        return -1
    n, m = len(circle), len(word)
    circle = circle * int(m//n + 2)


    for i in range(n):
        if circle[i:i+m] == word:
            return i, 1

    for i in range(n-1, -1, -1):
        if circle[i:i+m][::-1] == word:
            return (i+m-1) % n, -1
    return -1

find_word_in_circle('603561236667391524843', '248436035')
````

Написать функцию find_max_substring_occurrence(input_string), принимающую на вход непустую строку input_string. Функция должна возвращать наибольшее число k, такое что input_string совпадает с некоторой своей подстрокой t, выписанной k раз подряд.

Формат ввода

find_max_substring_occurrence(’abab’)

Формат вывода

2

Примечания

На проверку необходимо отправить файл с функцией find_max_substring_occurrence(input_string)

````python
def find_max_substring_occurrence(input_string):
    n = len(input_string)
    for k in range(n, 0, -1):
        if n % k == 0:
            if input_string[:n//k] * k == input_string:
                return k


find_max_substring_occurrence('aaaaaaaaa')
````

В файле input_dict_name находится человеко-драконий словарь. В первой строке словаря записано число слов, к которым есть перевод. Затем на каждой строке словаря располагается слов и один или несколько переводов к нему. Слово и его переводы разделены дефисом, переводы одного слова разделены запятой. Дефис отделён от соседних слов пробельными символами. После запятой ставится пробельный символ. Функция get_new_dictionary(input_dict_name, output_dict_name) должна по человеко-драконьему словарю, находящемуся в input_dict_name, построить драконе-человечий словарь и сохранить его в файл с именем output_dict_name в аналогичном исходному словарю формате. Словарь должен быть полным, т.е. учитывать всю информацию, которая находилась в исходном словаре. Строки выходного словаря должны быть отсортированы в лексикографическом порядке.

Формат ввода
Файл ’input.txt’:

5
cat - kosha
dog - soba
good - horo, normo
bad - ploh, uzha
ugly - uzha

Формат вывода
Файл ’output.txt’:

6
horo - good
kosha - cat
normo - good
ploh - bad
soba - dog
uzha - bad, ugly

Примечания

На проверку необходимо отправить файл с функцией get_new_dictionary(input_dict_name, output_dict_name).

````python
def get_new_dictionary(input_dict_name, output_dict_name):
    dr_to_hum = {}
    hum_to_dr = {}
    with open(input_dict_name, 'r') as file, open(output_dict_name, 'w') as out:
        n = int(file.readline().strip())
        for line in file:
            line = line.strip().split(' - ')
            dr_to_hum[line[0]] = line[1].split(', ')

        for key, value in dr_to_hum.items():
            for val in value:
                if val in hum_to_dr.keys():
                    hum_to_dr[val].append(key)
                else:
                    hum_to_dr[val] = [key]

        hum_to_dr = dict(sorted(hum_to_dr.items(), key=lambda item: item[0]))

        print(len(hum_to_dr.keys()), file = out)

        for key, value in hum_to_dr.items():
            value = list(sorted(value))
            print(f'{key} - {", ".join(value)}', file = out)
    file.close(), out.close()


get_new_dictionary('input.txt', 'noutput.txt')
````

Напишите функцию check_first_sentence_is_second, принимающую на вход две строки. Каждая строка задаёт предложение. Необходимо проверить, можно ли получить второе предложения из первого с помощью перестановки и удаления слов. Каждая из строк содержит только буквы и пробелы, любые буквенные последовательности разделённые пробелом считаются разными словами. Если можно, функция должна вернуть True, иначе False.

Формат ввода

check_first_sentence_is_second("люк я твой отец", "я отец твой")

Формат вывода

True

Примечания

На проверку необходимо отправить файл с функцией check_first_sentence_is_second.

````python
def check_first_sentence_is_second(first, second):
    words_dict = {}

    for word in first.split():
        if word in words_dict.keys():
            words_dict[word] += 1
        else:
            words_dict[word] = 1

    for word in second.split():
        if word in words_dict.keys() and words_dict[word] > 0:
            words_dict[word] -= 1
        else:
            return False

    return True

check_first_sentence_is_second("люк я твой отец", "я отец твой твой")
````

Необходимо написать нерекурсивную функцию find_path_sums, которая принимает на вход бинарное дерево (не обязательно сбалансированное), закодированное описанным способом в виде кортежа, производит его обход в глубину и для каждой листовой вершины вычисляет сумму всех значений от корня до неё включительно. Функция ничего не возвращает, вместо этого она выводит получаемые суммы в порядке следования листовых вершин (слева направо).

Формат ввода

find_path_sums((1, (2, None, None), (3, None, None)))

Формат вывода

3

4

````python
def find_path_sums(path):

    node = path

    stack = [(node, node[0])]

    while stack:
        node, sum = stack.pop()
        value, left, right = node

        if not left and not right:
            print(sum)

        else:
            if right:
                stack.append((right, sum + right[0]))
            if left:
                stack.append((left, sum + left[0]))

find_path_sums((1, (2, None, None), (3, None, None)))
````
