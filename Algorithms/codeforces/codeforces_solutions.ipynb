```python
# C
N = int(input())

for _ in range(N):
    n, k, p = list(map(int, input().split()))
    if k > n*p or -n*p > k:
        print(-1)
    elif k >= 0:
        print(k // p + 1 if k%p != 0 else k//p)
    else:
        print(-k // p + 1 if k%p != 0 else -k//p)
```

```python
# F
t = int(input())

for _ in range(t):
    n = int(input())
    arr = list(map(int, input().split()))
    ans = 'YES'
    for i in range(n//2):
        if arr[i] < 2*(n - i) - 1 or arr[n-i-1] < 2*(n-i) - 1:
            ans = 'NO'
            break
    if n%2 == 1:
        if arr[n//2] < n:
            ans = 'NO'
    print(ans)
```

```python
# I
t = int(input())

for _ in range(t):
    n, k =list( map(int, input().split()))
    arr = list(map(int, input().split()))

    cnt = [0]*(n+1)
    for x in arr:
        cnt[x] += 1

    score = 0

    for x in range(1, n+1):
        y = k - x

        if y < 1 or y > n:
            continue

        if x < y:
            score += min(cnt[x], cnt[y])
        elif x == y:
            score += cnt[x] // 2

    print(score)
```

```python
# L
import sys

t = int(input())

for _ in range(t):
    n = int(input())
    a = list(map(int, input().split()))

    freq = [0]*(n+1)
    for x in a:
        freq[x] += 1

    best_l = -1
    best_r = -1
    best_len = 0

    i = 0
    while i < n:
        if freq[a[i]] != 1:
            i += 1
            continue

        j = i
        while j < n and freq[a[j]] == 1:
            j += 1

        if j - i > best_len:
            best_len = j - i
            best_l = i
            best_r = j - 1

        i = j

    if best_len == 0:
        print(0)
    else:
        print(best_l + 1, best_r + 1)
```

```python
# N
t = int(input())

for _ in range(t):
    n, m = list(map(int, input().split()))

    arr = [list(map(int, input().split())) for p in range(n)]

    colors_1 = set()
    colors_2 = set()
    for i in range(n):
        for j in range(m):
            if arr[i][j] in colors_2:
                continue

            elif (j+1 <= m-1 and arr[i][j] == arr[i][j+1]) or (i+1 <= n-1 and arr[i][j] == arr[i+1][j]):
                colors_2.add(arr[i][j])
                if arr[i][j] in colors_1:
                    colors_1.remove(arr[i][j])
            else:
                colors_1.add(arr[i][j])

    if len(colors_2) > 0:
        print(len(colors_1) + (len(colors_2) - 1)*2)
    else:
        print(len(colors_1) - 1 + len(colors_2)*2)
```

```python
# O
t = int(input())

for _ in range(t):
    n, m = list(map(int, input().split()))
    arr = []

    for i in range(n):
        arr.append(list(map(int, input().split())))

    for i in range(n):
        row = []
        for j in range(m):
            if i - 1 >= 0:
                a = arr[i-1][j]
            else:
                a = float('-inf')

            if i + 1 <= n - 1:
                b = arr[i+1][j]
            else:
                b = float('-inf')

            if j - 1 >= 0:
                c = arr[i][j-1]
            else:
                c = float('-inf')

            if j + 1 <= m-1:
                d = arr[i][j+1]
            else:
                d = float('-inf')
            row.append(str(min(arr[i][j], max(a, b, c, d))))
        print(' '.join(row))
```

```python
# P
def get_answer(n, c, d, arr):
    a = arr[0]
    square = []

    for i in range(n):
        for j in range(n):
            square.append(a + i*c + j*d)

    square.sort()

    for i in range(n**2):
        if arr[i] != square[i]:
            return 'NO'
            break

    return 'YES'


t = int(input())

for _ in range(t):
    n, c, d = list(map(int, input().split()))
    arr = list(map(int, input().split()))
    arr.sort()
    print(get_answer(n, c, d, arr))
```

```python
# Q
t = int(input())

for _ in range(t):
    n = int(input())
    a = list(map(int, input().split()))

    arr = sorted([(a[i], i) for i in range(n)])

    vals = [x[0] for x in arr]

    pref = [0]*n
    pref[0] = vals[0]
    for i in range(1,n):
        pref[i] = pref[i-1] + vals[i]

    reach = [0]*n
    reach[-1] = n-1

    for i in range(n-2,-1,-1):
        if pref[i] >= vals[i+1]:
            reach[i] = reach[i+1]
        else:
            reach[i] = i

    ans = [0]*n

    for i in range(n):
        original_idx = arr[i][1]
        ans[original_idx] = reach[i]

    print(*ans)
```

```python
# R
def find_y(x):
    if x <= 2:
        return -1
    y = x - 1
    if y > (x & y) > 0:
        return y
    else:
        highest_bit = x.bit_length() - 1
        y = (1 << highest_bit) - 1
        if y > (x & y) > 0:
            return y
    return -1

t = int(input())

for _ in range(t):
    x = int(input())

    print(find_y(x))
```
