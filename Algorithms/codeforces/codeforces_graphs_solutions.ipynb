```python
# C
n, d = list(map(int, input().split()))
pod = input()

curr_lilly = 0
cand_lilly = None
totals = 1
for i in range(1, n):
    if i - curr_lilly > d:
        if cand_lilly and i - cand_lilly <= d:
            totals += 1
            curr_lilly = int(cand_lilly)

        else:
            totals = -1
            break

    if pod[i] == '1':
        cand_lilly = i

print(totals)
```

```python
# D
a, b = map(int, input().split())

from collections import deque

Queue = deque()
Queue.append((a, ''))

seen = set()
final_actions = None
while Queue:
    element, actions = Queue.popleft()
    if element == b:
        final_actions = actions
    if element < b and element not in seen:
        Queue.append((10 * element + 1, actions+ '1'))
        Queue.append((2 * element, actions+ '2'))

if not final_actions:
    print('NO')
else:
    print('YES')
    numbers = [str(a)]
    for action in final_actions:
        if action == '1':
            a = 10 * a + 1
        else:
            a = 2*a
        numbers.append(str(a))
    print(len(numbers))
    print(' '.join(numbers))
```

```python
# F

n, s = map(int, input().split())
s = s-1
row1 = input().split()
row2 = input().split()

has_intersect = False

for i in range(s, n):
    if row1[i] == row2[i] == '1':
        has_intersect = 1
        break

if row1[0] == '0':
    print('NO')
elif row1[s] == '1':
    print('YES')
elif row2[s] == '1' and has_intersect == 1:
    print('YES')
else:
    print('NO')
```

```python
# G

n = int(input())
states = list(map(int, input().split()))
states = [x-1 for x in states]

def travel_states(start):
    seen = set()
    i = start
    while i not in seen:
        seen.add(i)
        i = states[i]
    return i+1

output = []
for student in range(n):
    output.append(str(travel_states(student)))

print(' '.join(output))
```

```python
# H
R, C = map(int, input().split())

field = arr = [[x for x in input()] for p in range(R)]

def get_around(r, c):
    directions = [(0,1), (0,-1), (1,0), (-1,0)]
    result = []

    for dr, dc in directions:
        nr, nc = r + dr, c + dc
        if 0 <= nr < R and 0 <= nc < C:
            result.append((nr, nc))

    return result


def protect_filed(field):
    for r in range(R):
        for c in range(C):
            if field[r][c] == 'S':
                nearest = get_around(r, c)

                for i, j in nearest:
                    if field[i][j] == 'W':
                        return 'No'
                    if field[i][j] == '.':
                        field[i][j] = 'D'

    return 'Yes'

ans = protect_filed(field)

if ans == 'No':
    print('No')
else:
    print('Yes')

    for row in field:
        print(''.join(row))
```

```python
# J
n, t = map(int, input().split())
t = t-1

# portals
states = {i: x+i for i, x in enumerate(list(map(int, input().split())))}

x = 0
ans = 'NO'

while x < t:
    x = states[x]
    if x == t:
        ans = 'YES'
        break

print(ans)
```

```python
# K
n = int(input())

states = {}

for i in range(n):
    states[i] = int(input()) - 1


global_max = 0
for x in range(n):
    curr_max = 0
    curr_x = x
    while curr_x != -2:
        curr_max += 1
        curr_x = states[curr_x]
    if curr_max > global_max:
        global_max = curr_max

print(global_max)
```
