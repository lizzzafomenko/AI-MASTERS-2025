```python
# A
t = int(input())
import math

for _ in range(t):
    year = int(input())
    root = math.isqrt(year)
    if root * root == year:
        print(0, root)
    else:
        print(-1)
```

```python
# C
t = int(input())

for _ in range(t):
    n, x = map(int, input().split())
    arr = list(map(int, input().split()))

    arr.sort(reverse=True)

    cnt = 0
    teams = 0

    for a in arr:
        cnt += 1
        if cnt * a >= x:
            teams += 1
            cnt = 0

    print(teams)
```

```python
# F
n = int(input())
arr = list(map(int, input().split()))

measures = []

curr_num = 0
prev_num = 0
prev = None

for x in arr:
    if x == prev:
        curr_num += 1
    else:
        measures.append(min(curr_num, prev_num)*2)
        prev_num = curr_num
        curr_num = 1
    prev = x

    measures.append(min(curr_num, prev_num)*2)
print(max(measures))
```

```python
# G
n = int(input())

ax, ay = map(int, input().split())
bx, by = map(int, input().split())
cx, cy = map(int, input().split())

dx1 = bx - ax
dy1 = by - ay
dx2 = cx - ax
dy2 = cy - ay

if ((dx1 > 0 and dx2 > 0) or (dx1 < 0 and dx2 < 0)) and ((dy1 > 0 and dy2 > 0) or (dy1 < 0 and dy2 < 0)):
    print("YES")
else:
    print("NO")
```

```python
# H
n = int(input())
states = list(map(int, input().split()))
states = [x-1 for x in states]

def travel_states(start):
    seen = set()
    i = start
    while i not in seen:
        seen.add(i)
        i = states[i]
    return i+1

output = []
for student in range(n):
    output.append(str(travel_states(student)))

print(' '.join(output))
```

```python
# I
t = int(input())

for _ in range(t):
    n, k = list(map(int, input().split()))
    arr = list(map(int, input().split()))

    prev = None
    curr_days = 0
    total = 0

    for i in range(n):
        if arr[i] == 0:
            if prev == 0:
                curr_days += 1

            else:
                curr_days = 1

        else:
            total += (curr_days + 1) // (k + 1)
            curr_days = 0

        prev = arr[i]
    total += (curr_days + 1) // (k + 1)
    print(total)
```

```python
# K
smth = {a: i for i, a in enumerate('abcdefgh')}

s = input().strip()
t = input().strip()

x1, y1 = smth[s[0]], int(s[1])
x2, y2 = smth[t[0]], int(t[1])

dx = x2 - x1
dy = y2 - y1

moves = []

while dx != 0 or dy != 0:
    move = ''

    if dx > 0:
        move += 'R'
        dx -= 1
    elif dx < 0:
        move += 'L'
        dx += 1

    if dy > 0:
        move += 'U'
        dy -= 1
    elif dy < 0:
        move += 'D'
        dy += 1

    moves.append(move)

print(len(moves))
for m in moves:
    print(m)
```

```python
# M
a, b = map(int, input().split())

from collections import deque

Queue = deque()
Queue.append((a, ''))

seen = set()
final_actions = None
while Queue:
    element, actions = Queue.popleft()
    if element == b:
        final_actions = actions
    if element < b and element not in seen:
        Queue.append((10 * element + 1, actions+ '1'))
        Queue.append((2 * element, actions+ '2'))

if not final_actions:
    print('NO')
else:
    print('YES')
    numbers = [str(a)]
    for action in final_actions:
        if action == '1':
            a = 10 * a + 1
        else:
            a = 2*a
        numbers.append(str(a))
    print(len(numbers))
    print(' '.join(numbers))
```
