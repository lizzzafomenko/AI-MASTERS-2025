```python
class SLL_Node:
    def __init__(self, key, value):
        self.next = None
        self.key = key
        self.value = value

class SinglyLinkedList:
    def __init__(self):
        self.root = None
        self.last = None
        self.length = 0

    def add(self, key, value, ind = None, one_indexed = False):
        if isinstance(ind, int):
            return self.add_k(key, value, ind, one_indexed)

        node = SLL_Node(key, value)
        if self.root is None:
            self.root = node
            self.last = node
            self.length += 1
            return

        self.last.next = node
        self.last = node
        self.length += 1
        return

    def add_many(self, arr):
        for (key, value) in arr:
            self.add(key, value)

    def __repr__(self, print_all=True):

        representation = []

        node = self.root

        if node is None:
            return 'list is empty'

        else:
            for i in range(self.length):
                if print_all:
                    representation.append(f"({node.key} : {node.value})")
                else:
                    representation.append(f"{node.value}")
                node = node.next
        return ' -> '.join(representation) + f'\nlength = {self.length}\n'


    def change_k(self, ind, val, one_indexed = False):
        if one_indexed:
            ind -= 1

        node = self.root

        for curr_ind in range(self.length):
            if curr_ind == ind:
                node.value = val
            node = node.next

        return f'List consists of {curr_ind} nodes'


    def add_k(self, key, value, ind, one_indexed = False):
        assert ind <= self.length, f'List length is {self.length}, can not add node number {ind}'
        new_node = SLL_Node(key, value)
        if one_indexed:
            ind -= 1

        if ind == self.length+1:                  # tail case
            self.last.next = new_node
            self.last = new_node
            self.length += 1
            return

        if ind == 0:  # root case
            new_node.next = self.root
            self.root = new_node
            self.length += 1
            return

        node = self.root
        prev = None

        for curr_ind in range(self.length):

            if curr_ind == ind-1:
                parent = node
                next_node = node.next

                parent.next = new_node
                new_node.next = next_node
                self.length += 1
                return

            prev = node
            node = node.next


    def delete_k(self, ind, one_indexed = False):
        if ind == 'last':
            ind = self.length - 1

        elif ind == 'first':
            self.root = self.root.next
            self.length -= 1
            return

        assert ind <= self.length+1, f'List length is {self.length}, can not delete node number {ind}'

        if one_indexed:
            ind = -1

        node = self.root
        prev = None

        for curr_ind in range(self.length):

            if curr_ind == ind - 1:
                parent = node
                node_to_delete = node.next
                if curr_ind == self.length - 1:
                    self.last = parent
                    parent.next = None
                    return
                next_model = node.next.next     # тут может сломаться

                parent.next = next_model
                self.length -= 1
                return

            prev = node
            node = node.next

    def compress_list(self): # если одинаковое -> нужно
        prev_key = self.root.key
        node = self.root.next
        prev_node = self.root

        for _ in range(1, self.length):
            if node.key == prev_key:
                prev_node.next = node.next
                self.length -= 1
            else:
                prev_key = node.key
                prev_node = node
            node = node.next

        return

    def find_cycle(self):
        node = self.root

        for _ in range(self.length):
            node = node.next

        if node:  # если у нас нет цикла, то мы должны дойти до конца, до того узла, у которого не сущесвует
            return 'there is a cycle in the list'

        return 'no cycle found :)'




```

```python
# посмотрим, как работает дефолтный список

nodes_to_add = [('bayo', 'yayo'), ('ogo?', 'aga!'), ('its', 'me')]

sll = SinglyLinkedList()

# добавим много узлов
sll.add_many(nodes_to_add)
print('added nodes')
print(sll, '\n')

# удалим последний
sll.delete_k('last')
print('delete last element')
print(sll, '\n')

# и вставим такой же узел в начало...
sll.add_k('its', 'me', 0)
print('inserted node at the start')
print(sll, '\n')

# и снова уберем его...
sll.delete_k('first')
print('removing first element')
print(sll, '\n')

# и снова добавим в серединку
sll.add_k('its', 'me', 1)
print('add node at the middle')
print(sll, '\n')

# и снова удалим...
sll.delete_k(1)
print('remove node from the middle')
print(sll, '\n')

sll.change_k(1, 'oh my god')
print(sll, '\n')
```

Реализуйте двусвязный список в виде класса Double_linked_list. Добавьте храние не только первого элемента, но еще и последнего. Протестируйте функционал двусвязного списка. Реализуйте итеративную распечатку в прямом и обратном порядке.

```python
class DLL_Node:
    def __init__(self, key, value):
        self.parent = None
        self.next = None
        self.key = key
        self.value = value


class DoublyLinkedList:
    def __init__(self):
        self.root = None
        self.last = None
        self.length = 0

    def add(self, key, value, ind = None, one_indexed = False):
        if isinstance(ind, int):
            return self.add_k(key, value, ind, one_indexed)

        node = DLL_Node(key, value)
        if self.root is None:
            self.root = node
            self.last = node
            self.length += 1
            return

        self.last.next = node
        node.parent = self.last
        self.last = node
        self.length += 1
        return

    def add_many(self, arr):
        for (key, value) in arr:
            self.add(key, value)

    def __repr__(self, print_all=True):

        representation = []

        if self.length == 0:
            return 'list is empty'

        node = self.last
        for i in range(self.length):
            representation.append(f"({node.key} : {node.value})")
            node = node.parent
        ans = 'reverse (tail to root)\n' + ' -> '.join(representation) + f'\nlength = {self.length}\n'

        representation = []
        node = self.root
        for i in range(self.length):
            representation.append(f"({node.key} : {node.value})")
            node = node.next
        return 'forward (root to tail)\n' + ' -> '.join(representation) +'\n' + ans




    def change_k(self, ind, val, one_indexed = False):
        if one_indexed:
            ind -= 1

        node = self.root

        for curr_ind in range(self.length):
            if curr_ind == ind:
                node.value = val
            node = node.next

        return f'List consists of {curr_ind} nodes'


    def add_k(self, key, value, ind, one_indexed = False):
        assert ind <= self.length, f'List length is {self.length}, can not add node number {ind}'
        new_node = DLL_Node(key, value)
        if one_indexed:
            ind -= 1

        if ind == self.length+1:                  # tail case
            self.last.next = new_node
            new_node.parent = self.last
            self.last = new_node
            self.length += 1
            return

        if ind == 0:                              # root case
            new_node.next = self.root
            self.root.parent = new_node
            self.root = new_node
            self.length += 1
            return

        node = self.root
        prev = None

        for curr_ind in range(self.length):

            if curr_ind == ind-1:
                parent = node
                next_node = node.next

                parent.next = new_node
                new_node.parent = parent

                new_node.next = next_node
                next_node.parent = new_node
                self.length += 1
                return

            prev = node
            node = node.next


    def delete_k(self, ind, one_indexed = False):
        if ind == 'last':
            self.last.parent.next = None
            self.last = self.last.parent
            self.length -= 1
            return


        elif ind == 'first':
            self.root.next.parent = None
            self.root = self.root.next
            self.length -= 1
            return

        assert ind <= self.length+1, f'List length is {self.length}, can not delete node number {ind}'

        if one_indexed:
            ind = -1

        node = self.root
        prev = None

        for curr_ind in range(self.length):

            if curr_ind == ind - 1:
                parent = node
                node_to_delete = node.next
                if curr_ind == self.length - 1:
                    parent.next = None
                    self.last = parent
                    return

                next_node = node.next.next

                parent.next = next_node
                next_node.parent = parent
                self.length -= 1
                return

            prev = node
            node = node.next

    def compress_list(self): # если одинаковое -> нужно
        prev_key = self.root.key
        node = self.root.next
        prev_node = self.root

        for _ in range(1, self.length):
            if node.key == prev_key:
                prev_node.next = node.next         # prev node -> node -> next        prev node <- node <- next
                node.next.parent = prev_node
                self.length -= 1
            else:
                prev_key = node.key
                prev_node = node
            node = node.next

        return

    def find_cycle(self):
        node = self.root

        for _ in range(self.length):
            node = node.next

        if node:  # если у нас нет цикла, то мы должны дойти до конца, до того узла, у которого не сущесвует
            return 'there is a cycle in the list'

        return 'no cycle found :)'




```
