### Задача 2

Реализуйте решение задачи о банкомате (представляется ли число в виде суммы некоторого количества копий некоторого набора).

```python
def ATM_dinprog(money, target):
    if target in money:
        return target

    money.sort()
    money.reverse()

    dynprog = [0 for i in range(target+1)]
    dynprog[0] = 1

    backprop = [0 for i in range(target+1)]

    for amount in range(1, target+1):
        for banknote in money:
            if dynprog[amount - banknote] and amount - banknote >= 0:
                dynprog[amount] = 1
                backprop[amount] = banknote
                break

    if dynprog[-1] == 1:
        count_money = {key: 0 for key in money}
        i = target
        while i != 0:
            count_money[backprop[i]] += 1
            i = i - backprop[i]

        return count_money

    return False

def funny_bank(count_money, target):
    if count_money:
        print(f'your target sum {target}$ can be changed as:')
        for key, value in count_money.items():
            if value > 0:
                print(f'\t{key}$ : {value}')
```

```python
# пример с лекции
funny_bank(ATM_dinprog([3,5], 13), 13)
```

### Задача 3

Добавьте в реализацию задачи о рюкзаке возможность нахождения не только стоимости оптимального набора, но и самого набора.

```python

def backpack(W, weights, costs):
    n = len(costs)
    C = [[0 for i in range(n+1)] for j in range(W+1)]

    for w in range(1, W+1):
        for i in range(1, n+1):
            if weights[i-1] <= w:
                C[w][i] = max(C[w - weights[i-1]][i-1] + costs[i-1], C[w][i-1])
            else:
                C[w][i] = C[w][i-1]

    items = []
    for i in range(n, 0, -1):
        if C[w][i] != C[w][i - 1]:
            items.append(i - 1)
            w -= weights[i - 1]

    items.reverse()
    return C, items
```

```python
total_weight = 6
weights = [4, 6, 2, 4]
costs = [2, 4, 6, 5]

C, items = backpack(total_weight, weights, costs)
for row in C:
    print(row)

print('\n')
print(f'total cost is {C[-1][-1]}')
print(f'taken items: {' '.join([str(x) for x in items])}')
```
