### Задача 1

Самостоятельно реализуйте нахождение расстояния Левенштейна для двух строк с помощью динамического программирования. Можно читать книжки, Википедию, гуглить, но нельзя пользоваться нейросетями и чужим кодом.

```python
import numpy as np

def Levenstein_distance(string1, string2, del_cost = 1, ins_cost = 1, replace_cost = 1):
    n, m = len(string1), len(string2)

    dist = np.zeros((n+1, m+1))
    dist[0, :] = np.arange(m+1)
    dist[:, 0] = np.arange(n+1)

    for i in range(1, n+1):
        for j in range(1, m+1):
            dist[i, j] = min(dist[i, j-1] + del_cost, dist[i-1, j] + ins_cost, dist[i-1, j-1] + replace_cost * int(string1[i-1] != string2[j-1]))


    #print('     ', '  '.join(string2))
    #print(' ', dist[0, :])
    #for i in range(1, n+1):
    #    print(string1[i-1], dist[i, :])
    #print('\n')
    return dist

dist = Levenstein_distance('аргумент', 'рудимент')  # пример для сравнения - первое что вылазит в гугле при "расстояние Левенштейна"
f'Levenstein distance is {dist[-1, -1]}'
```

Реализуйте генератор игр слов на основе задачи 1 и набора слов в файле *10000-russian-words-cyrillic-only.txt* из репозитория https://github.com/hingston/russian/tree/master (для удобства тот же файл лежит в репозитории курса и в классруме). Для этого нужно:

- для каждого слова найти наиболее близкое по Левенштейну, но другое слово. Формально, нужно найти слово с минимальным расстоянием, но строго больше 0
- наслаждаться результатом и присылать самые забористые перлы в чат

```python
all_words = open('10000-russian-words-cyrillic-only.txt', 'r').read().strip().split('\n')

def find_nearest(target_word):
    all_findings = []
    for word in all_words:
        dist = Levenstein_distance(target_word, word)
        if dist[-1, -1] == 1:
            return word
        if dist[-1, -1] != 0:
            all_findings.append([word, dist[-1, -1]])

    all_findings = sorted(all_findings, key = lambda x: x[1])

    return all_findings[0][0]



phrase = 'Филипп Киркоров заявил, что помог Болгарии впервые выиграть Евровидение'
new_phrase = ''

for word in phrase.split(' '):
    new_phrase = new_phrase + ' ' + find_nearest(word.lower())

print(f'первоначальная фраза: {phrase}')
print(f'получилась фраза:    {new_phrase}')
```
