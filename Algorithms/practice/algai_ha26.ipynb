Выведите решение динамическим программированием для задачи поиска максимальной общей подстроки двух строк. Реализуйте его на питоне

```python
import numpy as np

def get_largest_substring(str1, str2):
    n1 = len(str1)
    n2 = len(str2)
    arr = np.zeros((n1+1, n2+1))

    for i in range(n1):
        for j in range(n2):
            if str1[i] == str2[j]:
                arr[i+1, j+1] = arr[i, j] + 1

            else:
                arr[i+1, j+1] = 0

    return arr


def get_substring_as_str(arr, str1):
    max_length = int(arr.max())

    idxs = np.where(arr == max_length)
    ans = []

    for idx in idxs[0]:
        ans.append(str1[idx - max_length : idx])

    return max_length, ans

```

```python
str1, str2 = 'цитомегаловирус', 'ортомиксовирус'

x = get_largest_substring(str1, str2)
x
```

```python
get_substring_as_str(x, str1)
```
